```python
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display
```

torch → PyTorch backend (tensor computation)
StableDiffusionPipeline → full LDM pipeline wrapper
display → show generated image in Colab

```python
print("CUDA Available:",torch.cuda.is_available())
```

```python
model_id = 'runwayml/stable-diffusion-v1-5'

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)
```

```python
pipe = pipe.to('cuda')
```

```python
prompt = "A futuristic cyberpunk city at night, ultra realistic, 4k, neon lights"
```

```python
output = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=7.5
)

print(type(output))
print(output)
```

```python
image = output.images[0]

from IPython.display import display
display(image)
```
