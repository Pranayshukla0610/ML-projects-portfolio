```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import statsmodels.api as sm
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv(r'C:\\Users\\prana\\Documents\\Python data\\MARTECH.csv')
data
```

```python
data = data.drop(columns=['Gender', 'Secondary School Stream', 'Career track', 'Batch', 'Name', 'Registration No.', 'Attendance', 'SN'])
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer_work_exp = SimpleImputer(strategy='most_frequent')
data['Designation'] = imputer_work_exp.fit_transform(data[['Designation']])
```

```python
imputer_internship = SimpleImputer(strategy='constant', fill_value='NA')
data['Internship(Designation)'] = imputer_internship.fit_transform(data[['Internship(Designation)']])
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer_work_exp = SimpleImputer(strategy='mean')
data['Work experience'] = imputer_work_exp.fit_transform(data[['Work experience']])
```

```python
imputer_internship = SimpleImputer(strategy='mean')
data['Internship(Tenure)'] = imputer_internship.fit_transform(data[['Internship(Tenure)']])
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data = pd.get_dummies(data, columns=['Communication Ability','Interpersonal Effectiveness','Customer Focus','Problem Solving & Analytical Skills','Leadership Skils','Teamwork Orientation','Creativity Innovation','Under Graduation','Designation','Internship(Designation)'], drop_first=True)
```
