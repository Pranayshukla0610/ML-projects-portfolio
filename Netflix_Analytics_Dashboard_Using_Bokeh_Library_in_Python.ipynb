```python
import pandas as pd
import numpy as np

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import column, row, gridplot
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LinearAxis,
    Range1d,
    NumeralTickFormatter,
    Div
)
from bokeh.palettes import Category20
```

```python
output_notebook()
```

```python
data = pd.DataFrame({
    'Year':[2016,2017,2018,2019,2020,2021,2022],
    'Movies':[200,250,300,400,500,550,600],
    'TV Shows':[100,120,150,200,250,300,350],
    'Ratings':[7.1,7.3,7.5,7.4,7.6,7.8,8.0],
    'Countries':[20,25,30,40,50,55,60]
})

source = ColumnDataSource(data)
```

```python
title = Div(text="""
<div style="
background-color:#111827;
padding:20px;
border-radius:10px;
">

<h1 style="
color:white;
text-align:center;
font-size:38px;
">
NETFLIX ANALYTICS DASHBOARD
</h1>

</div>
""")
```

```python
total_movies = data['Movies'].sum()
total_tv = data['TV Shows'].sum()
avg_rating = round(data['Ratings'].mean(),2)
total_countries = data['Countries'].sum()

kpi1 = Div(text=f"""
<div style="
background-color:#EF4444;
padding:20px;
border-radius:12px;
text-align:center;
color:white;
width:250px;
">
<h2>Total Movies</h2>
<h1>{total_movies}</h1>
</div>
""")

kpi2 = Div(text=f"""
<div style="
background-color:#3B82F6;
padding:20px;
border-radius:12px;
text-align:center;
color:white;
width:250px;
">
<h2>Total TV Shows</h2>
<h1>{total_tv}</h1>
</div>
""")

kpi3 = Div(text=f"""
<div style="
background-color:#10B981;
padding:20px;
border-radius:12px;
text-align:center;
color:white;
width:250px;
">
<h2>Average Rating</h2>
<h1>{avg_rating}</h1>
</div>
""")

kpi4 = Div(text=f"""
<div style="
background-color:#F59E0B;
padding:20px;
border-radius:12px;
text-align:center;
color:white;
width:250px;
">
<h2>Total Countries</h2>
<h1>{total_countries}</h1>
</div>
""")
```

```python
p1 = figure(
    title="Movies Growth Over Years",
    width=600,
    height=400,
    tools="pan,wheel_zoom,box_zoom,reset,save",
    background_fill_color="#1F2937",
    border_fill_color="#111827"
)

p1.line(
    x='Year',
    y='Movies',
    source=source,
    line_width=4,
    color='red'
)

p1.circle(
    x='Year',
    y='Movies',
    source=source,
    size=12,
    color='white'
)

hover1 = HoverTool(
    tooltips=[
        ('Year','@Year'),
        ('Movies','@Movies')
    ]
)

p1.add_tools(hover1)

p1.title.text_color = "white"
p1.xaxis.axis_label_text_color = "white"
p1.yaxis.axis_label_text_color = "white"
p1.xaxis.major_label_text_color = "white"
p1.yaxis.major_label_text_color = "white"
```

```python
p2 = figure(
    title="TV Shows Analysis",
    x_range=[str(x) for x in data['Year']],
    width=600,
    height=400,
    background_fill_color="#1F2937",
    border_fill_color="#111827"
)

p2.vbar(
    x=[str(x) for x in data['Year']],
    top=data['TV Shows'],
    width=0.5,
    color="#3B82F6"
)

hover2 = HoverTool(
    tooltips=[
        ('Year','$x'),
        ('TV Shows','$y')
    ]
)

p2.add_tools(hover2)

p2.title.text_color = "white"
p2.xaxis.major_label_text_color = "white"
p2.yaxis.major_label_text_color = "white"
```

```python
p3 = figure(
    title="Ratings Trend",
    width=600,
    height=400,
    background_fill_color="#1F2937",
    border_fill_color="#111827"
)

p3.varea(
    x=data['Year'],
    y1=0,
    y2=data['Ratings'],
    fill_color="#10B981",
    fill_alpha=0.7
)

p3.line(
    x=data['Year'],
    y=data['Ratings'],
    line_width=3,
    color='white'
)

hover3 = HoverTool(
    tooltips=[
        ('Year','$x'),
        ('Rating','$y')
    ]
)

p3.add_tools(hover3)

p3.title.text_color = "white"
p3.xaxis.major_label_text_color = "white"
p3.yaxis.major_label_text_color = "white"
```

```python
from math import pi
import numpy as np
import pandas as pd

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook

output_notebook()

# Example values
movie_total = 5000
tv_total = 2000

# DataFrame
pie_data = pd.DataFrame({
    'Category': ['Movies', 'TV Shows'],
    'Value': [movie_total, tv_total],
    'Angle': [
        movie_total/(movie_total+tv_total)*2*pi,
        tv_total/(movie_total+tv_total)*2*pi
    ],
    'Color': ['#EF4444', '#3B82F6']
})

# Create angles
pie_data['start_angle'] = np.append(
    0,
    np.cumsum(pie_data['Angle'])[:-1]
)

pie_data['end_angle'] = np.cumsum(
    pie_data['Angle']
)

# Source
pie_source = ColumnDataSource(pie_data)

# Figure
p4 = figure(
    height=400,
    width=500,
    title="Content Distribution",
    toolbar_location=None,
    tools="hover",
    tooltips="@Category: @Value",
    background_fill_color="#1F2937",
    border_fill_color="#111827"
)

# Wedge chart
p4.wedge(
    x=0,
    y=1,
    radius=0.4,
    start_angle='start_angle',
    end_angle='end_angle',
    color='Color',
    legend_field='Category',
    source=pie_source
)

# Styling
p4.axis.visible = False
p4.grid.grid_line_color = None
p4.title.text_color = "white"

show(p4)
```

```python
dashboard = column(

    title,

    row(
        kpi1,
        kpi2,
        kpi3,
        kpi4
    ),

    row(
        p1,
        p2
    ),   # ← THIS COMMA WAS MISSING

    row(
        p3,
        p4
    )

)

show(dashboard)
```
