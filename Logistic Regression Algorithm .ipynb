```python
from sklearn.model_selection import train_test_split
X = df1.drop("HeartDisease",axis=1)
y = df1["HeartDisease"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=22)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
pred = LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
from sklearn.metrics import r2_score
r2_sq = LR.score(X_test, y_test)
r2_sq
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
mse
```

```python
VIF = 1/(1-r2_sq)
VIF
```

```python
f = plt.figure(figsize=(14, 5))
# Add a subplot to the figure (1 row, 2 columns, 1st subplot)
ax = f.add_subplot(121)
sns.scatterplot(x=y_test, y=pred, ax=ax, color="r")
ax.set_title("Check for linearilty Actual vs Predicted")

ax = f.add_subplot(122)
sns.distplot((y_test - pred), ax=ax, color="b")
#vertical line to the Axes at a specified x-position.
ax.axvline((y_test - pred).mean(), color="k", linestyle="--")
ax.set_title("Check for Residual normality & mean")
```
