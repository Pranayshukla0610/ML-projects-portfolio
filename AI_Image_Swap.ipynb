```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from insightface.app import FaceAnalysis
from insightface.model_zoo import get_model
```

```python
app = FaceAnalysis(name='buffalo_l')

app.prepare(
    ctx_id=0,
    det_size=(640, 640)
)
```

```python
!wget https://github.com/deepinsight/insightface/releases/download/v0.7/inswapper_128.onnx
```

```python
swapper = get_model(
    'inswapper_128.onnx',
    download=False
)
```

```python
uploaded = files.upload()
```

```python
source_img = cv2.imread('/content/HD-wallpaper-shahrukh-khan-srk-actor-hero.jpg')
target_img = cv2.imread('/content/salman-Khan-header-1.jpg')
```

```python
source_faces = app.get(source_img)
target_faces = app.get(target_img)
```

```python
print("Source Faces:", len(source_faces))
print("Target Faces:", len(target_faces))
```

```python
result = swapper.get(
    target_img,
    target_faces[0],
    source_faces[0],
    paste_back=True
)
```

```python
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(result_rgb)
plt.axis("off")
plt.title("AI Face Swap Output")
plt.show()
```
