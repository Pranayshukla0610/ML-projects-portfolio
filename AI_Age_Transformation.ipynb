```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files
```

```python
model_id = 'runwayml/stable-diffusion-v1-5'

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    model_id,
    torch_dtype = torch.float16
)

pipe = pipe.to('cuda')
```

```python
uploaded = files.upload()
```

```python
init_img = Image.open('/content/My Pic 3.jpg').convert('RGB')
```

```python
init_image = init_img.resize((512,512))
```

```python
plt.imshow(init_image)
plt.axis("off")
plt.title("Original Face")
plt.show()
```

```python
prompt = """
elderly human face,
realistic wrinkles,
gray hair,
aged skin,
cinematic portrait,
ultra realistic,
detailed facial texture,
professional photography
"""
```

```python
prompt = """
young human face,
smooth skin,
youthful appearance,
modern hairstyle,
realistic portrait,
cinematic lighting,
high detail
"""
```

```python
negative_prompt = """
blurry,
low quality,
extra eyes,
distorted face,
bad anatomy,
deformed skin
  """
```

```python
result = pipe(
    prompt=prompt,
    image=init_image,
    strength=0.7,
    guidance_scale=8.5,
    negative_prompt=negative_prompt,
    num_inference_steps=40
).images[0]
```

```python
plt.imshow(result)
plt.axis("off")
plt.title("AI Age Transformation")
plt.show()
```
