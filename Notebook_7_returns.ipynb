```python
import random
import pandas as pd
import numpy as np

from datetime import timedelta
```

```python
random.seed(42)
np.random.seed(42)
```

```python
order_details_df = pd.read_csv('/content/order_details (2).csv')
```

```python
orders_df = pd.read_csv('/content/orders (2).csv')
```

```python
print("Order Details :", order_details_df.shape)

print("Orders :", orders_df.shape)
```

```python
return_base = order_details_df.merge(
    orders_df[['order_id','order_date']],
    on = 'order_id',
    how = 'left'
)
```

```python
return_base.head()
```

```python
RETURN_RATE = 0.10
```

```python
number_of_returns = int(
    len(return_base)
    * RETURN_RATE
)

print(number_of_returns)
```

```python
returned_orders = return_base.sample(
    n = number_of_returns,
    random_state = 42
).copy()
```

```python
returned_orders.reset_index(

    drop=True,

    inplace=True

)
```

```python
returns_base = order_details_df.merge(

    orders_df[["order_id", "order_date"]],

    on="order_id",

    how="left"

)
```

```python
return_reasons = [

    "Damaged Product",

    "Wrong Item Delivered",

    "Customer Changed Mind",

    "Size Issue",

    "Quality Issue",

    "Late Delivery",

    "Defective Product",

    "Missing Parts",

    "Duplicate Order",

    "Other"

]
```

```python
reason_probability = [

    0.18,

    0.10,

    0.20,

    0.10,

    0.12,

    0.08,

    0.12,

    0.05,

    0.03,

    0.02

]
```

```python
return_status_options = [

    "Approved",

    "Rejected",

    "Pending"

]
```

```python
return_status_probability = [

    0.80,

    0.10,

    0.10

]
```

```python
inspection_status_options = [

    "Passed",

    "Failed",

    "Pending"
]
```

```python
inspection_probability = [

    0.75,

    0.10,

    0.15

]
```

```python
refund_methods = [

    "UPI",

    "Credit Card",

    "Debit Card",

    "Wallet",

    "Bank Transfer",

    "Store Credit"

]
```

```python
return_id_list = []

order_detail_id_list = []

return_date_list = []

return_reason_list = []

refund_amount_list = []

return_status_list = []

inspection_status_list = []

restocking_fee_list = []

refund_method_list = []
```

```python
return_counter = 1
```

```python
print("=" * 65)

print("RETURNS DATASET CONFIGURATION")

print("=" * 65)

print("Order Detail Records :", len(order_details_df))

print("Return Rate :", RETURN_RATE * 100, "%")

print("Expected Returns :", number_of_returns)

print("=" * 65)
```

```python
for _, row in returned_orders.iterrows():

    # Return ID
    return_id = f"RET{return_counter:06d}"
    return_counter += 1

    # Order Detail ID
    order_detail_id = row["order_detail_id"]

    # Order Date
    order_date = pd.to_datetime(row["order_date"])

    # Return Date (1–30 days after purchase)
    return_date = order_date + timedelta(
        days=random.randint(1, 30)
    )

    # Return Reason
    return_reason = np.random.choice(
        return_reasons,
        p=reason_probability
    )

    # Return Status
    return_status = np.random.choice(
        return_status_options,
        p=return_status_probability
    )

    # Inspection Status
    inspection_status = np.random.choice(
        inspection_status_options,
        p=inspection_probability
    )

    # Refund Method
    refund_method = random.choice(
        refund_methods
    )

    # Line Sales
    line_sales = row["line_sales"]

    # Restocking Fee
    if return_reason in [
        "Customer Changed Mind",
        "Duplicate Order"
    ]:

        restocking_fee = round(
            line_sales * random.uniform(0.05, 0.10),
            2
        )

    else:

        restocking_fee = 0

    # Refund Amount
    if return_status == "Approved":

        refund_amount = round(
            line_sales - restocking_fee,
            2
        )

    else:

        refund_amount = 0

    # Store Data

    return_id_list.append(return_id)

    order_detail_id_list.append(order_detail_id)

    return_date_list.append(return_date)

    return_reason_list.append(return_reason)

    refund_amount_list.append(refund_amount)

    return_status_list.append(return_status)

    inspection_status_list.append(inspection_status)

    restocking_fee_list.append(restocking_fee)

    refund_method_list.append(refund_method)
```

```python
returns_df = pd.DataFrame({

    "return_id": return_id_list,

    "order_detail_id": order_detail_id_list,

    "return_date": return_date_list,

    "return_reason": return_reason_list,

    "refund_amount": refund_amount_list,

    "return_status": return_status_list,

    "inspection_status": inspection_status_list,

    "restocking_fee": restocking_fee_list,

    "refund_method": refund_method_list

})
```
