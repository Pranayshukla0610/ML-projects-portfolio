```python
import sqlite3

conn = sqlite3.connect('/content/superstore.db')
cursor = conn.cursor()

# Customers
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers AS
SELECT DISTINCT
customer_id,
customer_name,
segment,
city,
state,
country
FROM sales
""")

# Products
cursor.execute("""
CREATE TABLE IF NOT EXISTS products AS
SELECT DISTINCT
product_id,
product_name,
category,
sub_category
FROM sales
""")

# Orders
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders AS
SELECT
order_id,
customer_id,
product_id,
sales,
quantity,
profit,
order_date
FROM sales
""")

conn.commit()
conn.close()

print("Tables Created Successfully")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('/content/superstore.db')
```

```python
query = """

SELECT

o.order_id,
c.customer_name,
p.product_name,
o.sales

FROM orders o

JOIN customers c
ON o.customer_id = c.customer_id

JOIN products p
ON o.product_id = p.product_id

LIMIT 20

"""

df = pd.read_sql_query(query, conn)

df
```

```python
query = """

SELECT

c.customer_name,
o.order_id,
o.sales

FROM customers c

LEFT JOIN orders o

ON c.customer_id = o.customer_id

LIMIT 20

"""

pd.read_sql_query(query, conn)
```

```python
#Aggregations

query = """

SELECT

customer_id,
SUM(sales) total_sales

FROM orders

GROUP BY customer_id

ORDER BY total_sales DESC

LIMIT 10

"""

pd.read_sql_query(query, conn)
```

```python
#CTE
query = """

WITH customer_sales AS

(

SELECT

customer_id,
SUM(sales) total_sales

FROM orders

GROUP BY customer_id

)

SELECT *

FROM customer_sales

WHERE total_sales > 10000

ORDER BY total_sales DESC

"""

pd.read_sql_query(query, conn)
```

```python
#Window Function - ROW_NUMBER

query = """

SELECT

customer_id,

SUM(sales) total_sales,

ROW_NUMBER()

OVER(

ORDER BY SUM(sales) DESC

)

AS row_num

FROM orders

GROUP BY customer_id

"""

pd.read_sql_query(query, conn)
```

```python
#Window Function - RANK

query = """

SELECT

customer_id,

SUM(sales) total_sales,

RANK()

OVER(

ORDER BY SUM(sales) DESC

)

AS ranking

FROM orders

GROUP BY customer_id

"""

pd.read_sql_query(query, conn)
```

```python
#DENSE_RANK

query = """

SELECT

customer_id,

SUM(sales) total_sales,

DENSE_RANK()

OVER(

ORDER BY SUM(sales) DESC

)

AS dense_rank

FROM orders

GROUP BY customer_id

"""

pd.read_sql_query(query, conn)
```

```python
#NTILE

query = """

SELECT

customer_id,

SUM(sales) total_sales,

NTILE(3)

OVER(

ORDER BY SUM(sales)

)

AS segment

FROM orders

GROUP BY customer_id

"""

pd.read_sql_query(query, conn)
```

```python
#Running Total
query = """

SELECT

order_date,

SUM(sales) daily_sales,

SUM(SUM(sales))

OVER(

ORDER BY order_date

)

AS running_total

FROM orders

GROUP BY order_date

"""

pd.read_sql_query(query, conn)
```
