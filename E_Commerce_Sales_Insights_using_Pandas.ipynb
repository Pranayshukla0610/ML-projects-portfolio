```python
df['CustomerID'].fillna(0,inplace=True)
```

```python
df.dropna(subset=['Description'],inplace=True)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.head()
```

```python
df['Total_Price'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Country'] = df['Country'].str.lower()
```

```python
df.head()
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['WeekDay'] = df['InvoiceDate'].dt.weekday
```

```python
df.head()
```

```python
df['IsBulkOrder'] = df['Quantity'].apply(lambda x : 1 if x > 50 else 0)
df['HighValue'] = df['Total_Price'] > df['Total_Price'].mean()
```

```python
df.head()
```

```python
df.groupby('Country')['Quantity'].sum()
df.groupby('Month')['Total_Price'].mean()
df.groupby(['Country','Month'])['Total_Price'].sum()
```

```python
df.head()
```

```python
pivot = pd.pivot_table(df, values='Total_Price', index='Month', columns='Country', aggfunc='sum')
pivot.head()
```

```python
customers = df[['CustomerID','Country']].drop_duplicates()
merged = pd.merge(df, customers, on='CustomerID', how='left')
```

```python
df['RollingAvg'] = df['Total_Price'].apply(lambda x : "High" if x > 100 else 'low')
```

```python
multi = df.groupby(['Country','Month']).sum(numeric_only=True)
multi.index
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(x = 'Month', y='Total_Price', data=df)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df['Month'],
    df['Quantity'],
    df['Total_Price']
)

ax.set_xlabel('Month')
ax.set_ylabel('Quantity')
ax.set_zlabel('Total_Price')
```
