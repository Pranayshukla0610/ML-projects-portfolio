```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from statsmodels.api import OLS, add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
import statsmodels.api as sm
```

```python
data = pd.read_csv(r'C:\\Users\\prana\\Documents\\Python data\\Sales & Services1.csv')
data
```

```python
data.info()
```

```python
# Replace 'NA' with 0, convert to numeric
data.replace('NA', np.nan, inplace=True)
data['Work_Ex'] = pd.to_numeric(data['Work_Ex'], errors='coerce').fillna(0)
data['Intern'] = pd.to_numeric(data['Intern'], errors='coerce').fillna(0)
```

```python
data = data.drop(columns=["CT","PE"])
data.head()
```

```python
categorical_columns = ['COM', 'PERS', 'LEAD', 'Cust Focus', 'Prob', 'TEAM', 'CREA']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
data
```

```python
y = data['Score']  # Replace with the actual dependent variable name
X = data.drop('Score', axis=1)
```

```python
X_const = add_constant(X)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Filter columns with VIF < 10
low_vif_features = vif_data[vif_data['VIF'] < 10]['Feature']
X = X[low_vif_features]
```

```python
from sklearn.preprocessing import StandardScaler, PowerTransformer
power_transformer = PowerTransformer()
X_transformed = power_transformer.fit_transform(X)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_const)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import RidgeCV, LassoCV
lasso = LassoCV(cv=5)
lasso.fit(X_train, y_train)
```

```python
ridge_model = RidgeCV(alphas=[0.1, 1.0, 10.0], cv=5)
ridge_model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)
best_rf = grid_search.best_estimator_
```

```python
y_pred = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.stats.outliers_influence import variance_inflation_factor
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
pmae = np.mean(np.abs((y_test - y_pred) / (y_test + 1e-10))) * 100  # Adding a small constant to prevent infinite values
```

```python
X_train_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train_const).fit()
r_squared = model.rsquared
adj_r_squared = model.rsquared_adj
f_statistic = model.fvalue
prob_f_stat = model.f_pvalue
dw_stat = sm.stats.stattools.durbin_watson(model.resid)
```

```python
# Jarque-Bera test for normality of residuals
from scipy import stats
jb_test = stats.jarque_bera(model.resid)
jarque_bera, prob_jb = jb_test[0], jb_test[1]
```

```python
print(f"R-squared: {r_squared}")
print(f"Adjusted R-squared: {adj_r_squared}")
print(f"F-statistic: {f_statistic}")
print(f"Prob (F-statistic): {prob_f_stat}")
print(f"Durbin-Watson: {dw_stat}")
print(f"Jarque-Bera: {jarque_bera}")
print(f"Prob (Jarque-Bera): {prob_jb}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Percentage Mean Absolute Error (PMAE): {pmae}")
```
