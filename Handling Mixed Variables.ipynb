```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'C:\\Users\\prana\\Documents\\Python data\\Titanic2.csv',usecols=['Cabin','Ticket','Number','Survived'])
data.head()
```

```python
data['Number'].unique()
```

```python
fig = data['Number'].value_counts().plot.bar()
fig.set_title('Passenger Travelling With')
```

```python
# Extract Numerical Part
data['number_numerical'] = pd.to_numeric(data['Number'], errors='coerce', downcast='integer')
```

```python
# Extract Categorical Part
data['number_categorical'] = np.where(data['number_numerical'].isnull(),data['Number'],np.nan)
data.head()
```

```python
data['Cabin'].unique()
```

```python
data['Ticket'].unique()
```

```python
data['cabin_num'] = data['Cabin'].str.extract('(\d+)') # captures numerical part
data['cabin_cat'] = data['Cabin'].str[0] # captures categorical data
```

```python
data.head()
```

```python
data['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# Extract the last bit of ticket as number
data['ticket_num'] = data['Ticket'].apply(lambda s:s.split()[-1])
data['ticket_num'] = pd.to_numeric(data['ticket_num'],errors='coerce', downcast='integer')

data['ticket_cat'] = data['Ticket'].apply(lambda s:s.split()[0])
data['ticket_cat'] = np.where(data['ticket_cat'].str.isdigit(), np.nan, data['ticket_cat'])
```
